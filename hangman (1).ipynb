```python
import random

words = ["apple", "banana", "cherries"]
word = random.choice(words)

print(word)
n = len(word)

display = ["*"] * n
guess = []
m = 6

while m > 0:

    print("Word:", "".join(display))

    choose = input("enter your alphabet: ").lower()

    if len(choose) != 1 or not choose.isalpha():
        print("enter a single valid alphabet")
        continue

    if choose in guess:
        print("you already used this letter")
        continue

    guess.append(choose)

    if choose in word:
        print("you are correct")

        for i in range(n):
            if choose == word[i]:
                display[i] = choose
    else:
        print("wrong choose")
        m = m - 1

    print("remaining attempts:", m)

    if "".join(display) == word:
        print("you won the game")
        break

if m == 0:
    print("you lost the game. the word is:", word)
```
